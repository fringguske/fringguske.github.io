# Sunglasses are one of the best perfoming niche in dropshipping industry 
Python language is used in the analysis and visualization of this project .Driven with the disire to generate insights from raw data ,i extracted really meaningfull info from amazon sunglasses data .

```python
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')

# Load the dataset
df = pd.read_csv('/kaggle/input/2024-best-selling-sunglasses-on-amazon/best_sellers_amazon_2024_sunglasses.csv')

# Display column names and data types
print(df.dtypes)
```

```python
# Count the number of unique products
unique_products = df['title'].nunique()
print("Number of unique sunglasses products:", unique_products)
```

```python
# Count the number of products per brand and select the top 5
top_brands = df['brand'].value_counts().head(5)
print("Top 5 brands with the highest number of products:")
print(top_brands)
```

```python
# Calculate the average price
average_price = df['price/value'].mean()
print("Average price of sunglasses:", average_price)
```

```python
# Find the highest and lowest prices and their corresponding products
highest_price = df[df['price/value'] == df['price/value'].max()]
lowest_price = df[df['price/value'] == df['price/value'].min()]

print("Highest price and corresponding product:")
print(highest_price[['title', 'price/value']])
print("\nLowest price and corresponding product:")
print(lowest_price[['title', 'price/value']])
```

```python
# Calculate the average rating
average_rating = df['stars'].mean()
print("Average rating of sunglasses:", average_rating)
```

```python
# Calculate the total count of reviews
total_reviews = df['reviewsCount'].sum()
print("Total count of reviews across all sunglasses products:", total_reviews)
```

```python
# Calculate the median price
median_price = df['price/value'].median()
print("Median price of sunglasses:", median_price)
```

```python
# Calculate the standard deviation of prices
price_std = df['price/value'].std()
print("Standard deviation of prices for sunglasses:", price_std)
```

```python
# Find the highest number of reviews
max_reviews = df['reviewsCount'].max()
print("Highest number of reviews received by a single sunglasses product:", max_reviews)
```

```python
# Calculate the mode of ratings
rating_mode = df['stars'].mode()[0]  # Mode can return multiple values, selecting the first one
print("Mode of the ratings for sunglasses:", rating_mode)
```

```python
# Calculate the range of prices
price_range = df['price/value'].max() - df['price/value'].min()
print("Range of prices for sunglasses:", price_range)
```

```python
# Calculate the 25th percentile (first quartile) of prices
price_25th_percentile = df['price/value'].quantile(0.25)
print("25th percentile of prices for sunglasses:", price_25th_percentile)
```

```python
# Calculate the skewness of prices
price_skewness = df['price/value'].skew()
print("Skewness of the distribution of prices for sunglasses:", price_skewness)
```

```python
# Calculate the kurtosis of prices
price_kurtosis = df['price/value'].kurtosis()
print("Kurtosis of the distribution of prices for sunglasses:", price_kurtosis)
```

```python
# Calculate the interquartile range (IQR) of prices
Q1 = df['price/value'].quantile(0.25)
Q3 = df['price/value'].quantile(0.75)
IQR = Q3 - Q1
print("Interquartile range (IQR) of prices for sunglasses:", IQR)
```

```python
# Calculate the correlation coefficient between prices and ratings
price_rating_correlation = df['price/value'].corr(df['stars'])
print("Correlation coefficient between prices and ratings for sunglasses:", price_rating_correlation)
```

```python
# Calculate the average number of reviews per product
average_reviews_per_product = df['reviewsCount'].mean()
print("Average number of reviews per product:", average_reviews_per_product)
```

```python
# Calculate the coefficient of variation (CV) of prices
price_cv = (df['price/value'].std() / df['price/value'].mean()) * 100
print("Coefficient of variation (CV) of prices for sunglasses:", price_cv)
```

```python
# Calculate the median number of reviews per product
median_reviews_per_product = df['reviewsCount'].median()
print("Median number of reviews per product:", median_reviews_per_product)
```

```python
# Calculate the percentage of products with ratings above 4.5 stars
high_rating_percentage = (df[df['stars'] > 4.5]['title'].count() / df['title'].count()) * 100
print("Percentage of products with ratings above 4.5 stars:", high_rating_percentage)
```

```python
# Calculate the ratio of the highest price to the lowest price
price_ratio = df['price/value'].max() / df['price/value'].min()
print("Ratio of the highest price to the lowest price among sunglasses:", price_ratio)
```

```python
# Calculate the mode of the pricing currency
currency_mode = df['price/currency'].mode()[0]  # Mode can return multiple values, selecting the first one
print("Mode of the pricing currency for sunglasses:", currency_mode)
```

```python
# Calculate the correlation coefficient between number of reviews and average ratings
reviews_rating_correlation = df['reviewsCount'].corr(df['stars'])
print("Correlation coefficient between number of reviews and average ratings for sunglasses:", reviews_rating_correlation)
```

```python
# Calculate the proportion of products with descriptions available
products_with_description = df['description'].notnull().sum()
total_products = df.shape[0]
proportion_with_description = (products_with_description / total_products) * 100
print("Proportion of sunglasses products with descriptions available:", proportion_with_description)
```

```python
# Calculate the median number of stars received per review
median_stars_per_review = df['stars'] / df['reviewsCount']
median_stars_per_review = median_stars_per_review.median()
print("Median number of stars received per review for sunglasses:", median_stars_per_review)
```

```python
# Calculate the percentage of products with ratings below 3 stars
low_rating_percentage = (df[df['stars'] < 3]['title'].count() / df['title'].count()) * 100
print("Percentage of sunglasses products with ratings below 3 stars:", low_rating_percentage)
```

**Histogram showing the count of sunglasses products per brand in the dataset**

**Histogram showing the distribution of prices for sunglasses products per currency used**
